```python
# Importação das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
# Carregamento da base de dados e visualização das dimensões
base = pd.read_csv('dados.csv', sep=';')
base.shape
```

```python
# Visualização das 5 primeiras linhas
base.head()
```

Podemos notar que se trata de uma base pequena, mas que servirá bem ao proposito do exercício que é de visualização de dados

```python
#gráfico de dispersão baseado no Pib dos municipios do dataset
srn.distplot(base.PIB, bins = 12, axlabel = 'PIB').set_title('PIB')
```

```python
# gráfico de dispersão valor do Empenho por municipios do dataset
srn.distplot(base.VALOREMPENHO, bins = 10, axlabel = 'Valor do Empenho').set_title('Empenho')
```

```python
# gráfico de barras selecionado apenas os 10 municipios com maior valor de Pib
agrupado = base.sort_values('PIB',ascending = False).head(10)
agrupado = agrupado.iloc[:,1:3]
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='PIB', color = 'gray')
```

```python
# gráfico de barras selecionado apenas os 10 municipios com maior valor de Empenho
agrupado = base.sort_values('VALOREMPENHO', ascending = False).head(10)
agrupado = agrupado.iloc[:,[1,3]]
agrupado
agrupado.plot.bar(x='MUNICIPIO',y='VALOREMPENHO', color = 'gray')
```

```python
# Neste gráfico de dispersão é possível observar que em nem todos os casos o Pib tem relação com o 
# valor do Empenho. Como podemos ver o municipio de MUITOS CAPOES tem um Pib alto, mas o valor do 
# Empenho é menor se comparado com outros municipios com Pib um pouco menor como MARAU. 

agrupado = base.sort_values('PIB', ascending = False).head(10)
agrupado
srn.scatterplot(x='VALOREMPENHO', y='PIB', data=agrupado, hue='MUNICIPIO')
```
